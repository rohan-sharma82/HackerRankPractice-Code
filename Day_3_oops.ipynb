```python
class Phone:
  def __init__(self,price,camera,brand):
    self.price=price
    self.camera=camera
    self.brand=brand

  def buy(self):
    print("Phone is purchased")

class Smartphone(Phone):

  def buy(self):
    print("New SmartPhone is purchased")
    super().buy()
```

```python
#Method Overriding
q= Smartphone(20000, 34, "Redmi")

q.buy()
```
