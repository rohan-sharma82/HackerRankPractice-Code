#QUESTION
The provided code stub will read in a dictionary containing key/value
pairs of name:[marks] for a list of students. Print the average of
the marks array for the student name provided, showing 2 places after the decimal.

```python
#MY MESSED UP VERSION

n = int(input())
student_marks = {}
for _ in range(n):
  name, *line = input().split()
  scores = list(map(float, line))
  student_marks[name] = scores
query_name = input()
print(student_marks)
marks= student_marks[query_name]
print(marks)
total_marks=0
for i in marks:
  total_marks= total_marks+i
average= total_marks/len(marks)
print(average)
```
