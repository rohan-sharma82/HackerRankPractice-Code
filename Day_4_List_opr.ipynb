#Logic decided by Me

```python
n=int(input())
number=[]
for _ in range(n):
  operation,*num= input().split()
  number=list(map(int,num))

  if operation=="append" or operation=="insert":
    num.append(number)

  elif operation=="print":
    print(number)

  elif operation=="remove":
    number.remove(num)

  elif operation=="sort":
    number = number.sort()

  elif operation=="pop":
    number.pop()

  elif operation=="reverse":
    number.reverse()

  else:
    print("Invalid Operation")

print(number)
```

```python

n = int(input())
number = []
for _ in range(n):
    parts = input().split()
    operation = parts[0]

    if operation == "insert":
        i, e = int(parts[1]), int(parts[2])
        number.insert(i, e)
    elif operation == "append":
        e = int(parts[1])
        number.append(e)
    elif operation == "print":
        print(number)
    elif operation == "remove":
        e = int(parts[1])
        number.remove(e)
    elif operation == "sort":
        number.sort()
    elif operation == "pop":
        number.pop()
    elif operation == "reverse":
        number.reverse()
```
